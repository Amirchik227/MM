```python
import matplotlib.pyplot as plt
a = 0.3#вер-ть того, что травоядные размножатся
b = 0.28#вер-ть того, что травоядное будет съедено
c = 0.7#вер-ть того, что хищник умрет от голода
d = 0.3#вер-ть того, что хищнику хватит еды на размножение
x = [10]#численность жертв
y = [5]#численность хищников
dt = 0.02
t = [0]
dx, dy = [], []
for i in range(20000):
  dx.append((a - b * y[-1]) * x[-1])
  dy.append((-c + d * x[-1]) * y[-1])
  x.append(x[-1] + dx[-1] * dt)
  y.append(y[-1] + dy[-1] * dt)
  t.append(t[-1] + dt)
#print(len(t), len(x), len(y))
plt.plot(x,y, color='orange')
plt.xlabel('Preys')
plt.ylabel('Predators')
plt.grid()
```

```python
plt.plot(t, y, color='red', label='Predators')
plt.plot(t, x, color='blue', label='Preys')
plt.xlabel('t')
plt.ylabel('Population')
plt.grid()
plt.legend()
plt.show()
```
